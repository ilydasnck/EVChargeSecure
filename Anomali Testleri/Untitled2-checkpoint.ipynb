```python
# ===============================
# 1 Ktphaneler
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ruptures as rpt
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

# ===============================
# 2 Veri Ykleme
# ===============================
file_path = r"C:\Users\Mahmut\Downloads\UrbanEVDataset\UrbanEVDataset\20220901-20230228_zone-cleaned-aggregated\adj.csv"
df = pd.read_csv(file_path)

# Kolonlar kontrol et
print("Kolonlar:", df.columns)

# rnek olarak '102' kolonunu kullanyoruz
df['102'] = pd.to_numeric(df['102'], errors='coerce')

# ===============================
# 3 Index ayarlama
# ===============================
df.reset_index(drop=True, inplace=True)

# ===============================
# 4 Basit grselletirme
# ===============================
plt.figure(figsize=(15,5))
plt.plot(df['102'], label='arj lm')
plt.xlabel('lm Sras')
plt.ylabel('Deer')
plt.title('arj lm Deerleri - Zaman Serisi')
plt.legend()
plt.show()

# ===============================
# 5 Changepoint Detection (ruptures)
# ===============================
signal = df['102'].values
algo = rpt.Pelt(model="l2").fit(signal)
result = algo.predict(pen=10)

plt.figure(figsize=(15,5))
plt.plot(signal, label='arj lm')
for i, cp in enumerate(result):
    plt.axvline(cp, color='r', linestyle='--', label='Changepoint' if i==0 else "")
plt.xlabel('lm Sras')
plt.ylabel('Deer')
plt.title('Changepoint Detection - arj Profili')
plt.legend()
plt.show()

# ===============================
# 6 Isolation Forest ile Anomali Tespiti
# ===============================
from pyod.models.iforest import IForest

clf = IForest()
clf.fit(df[['102']])

df['iso_anomaly'] = clf.labels_  # 1 = anomali, 0 = normal

plt.figure(figsize=(15,4))
plt.plot(df['102'])
plt.scatter(df.index[df['iso_anomaly']==1],
            df['102'][df['iso_anomaly']==1],
            color='red')
plt.title("Isolation Forest Anomali Tespiti - Kolon 102")
plt.show()


# ===============================
# 7 LOF (Local Outlier Factor) ile Anomali Tespiti
# ===============================
# n_neighbors artrld, duplicate uyars azaltld
from pyod.models.lof import LOF

lof = LOF()
lof.fit(df[['102']])

df['lof_anomaly'] = lof.labels_

plt.figure(figsize=(15,4))
plt.plot(df['102'])
plt.scatter(df.index[df['lof_anomaly']==1],
            df['102'][df['lof_anomaly']==1],
            color='orange')
plt.title("LOF Anomali Tespiti - Kolon 102")
plt.show()


# ===============================
# 8 zet: Anomalileri iaretleme
# ===============================
# Hem IF hem LOF tarafndan tespit edilen anomalileri iaretle
df['final_anomaly'] = ((df['iso_anomaly']==1) | (df['lof_anomaly']==1)).astype(int)
df['final_anomaly'].value_counts()
```
